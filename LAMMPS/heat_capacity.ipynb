```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('step_toteng_NMC622_400K.csv')

# To display the first few rows of the DataFrame
print(df.head())
```

```python
energy = df[df['Step'].astype(int) > 1000]['TotEng']
```

```python
print(energy)
```

```python
number_of_simulation_box = 8
```

```python
#Unit_cell_mass in kilogam
unit_cell_mass = (15*6.941 + 9*58.6934 + 30*15.999 + 3*54.938 + 3*58.9332)*number_of_simulation_box*(1.66e-27)
print(unit_cell_mass)
```

```python
import numpy as np
```

```python
import numpy as np
# Constants
k_B = 8.617333262145e-5  # Boltzmann constant in eV/K
temperature = 400  # K, constant in NVT


# Compute fluctuations
E_avg = np.mean(energy)
E2_avg = np.mean(energy**2)
fluctuation = E2_avg - E_avg**2

# Estimate C_V
C_v = fluctuation / (k_B * temperature**2)  # units: eV/K
#C_v in J/kg.K
CV = (C_v*1.6e-19)/(unit_cell_mass)
print(CV) # Thermal capacity of the unit cell.
```
